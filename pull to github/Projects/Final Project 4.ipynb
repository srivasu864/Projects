```python
#RNN
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM,Dropout

df=pd.read_csv('C:\\vas.12344\\Google_Stock_Price_Train.csv')
df
f11=df.iloc[:,1:2]
f11
s1=MinMaxScaler()
f11=s1.fit_transform(f11)
f11.shape
x_train=[]
y_train=[]
for i in range(60,1258):
   x_train.append(f11[i-60:i,0])
   y_train.append(f11[i,0])
x_train,y_train=np.array(x_train),np.array(y_train)
x_train=x_train.reshape(1198,60,1)
m=Sequential()
m.add(LSTM(units=50,return_sequences=True,input_shape=(60,1)))
m.add(Dropout(0.2))
m.add(LSTM(units=50,return_sequences=True))
m.add(Dropout(0.2))
m.add(LSTM(units=50,return_sequences=True))
m.add(Dropout(0.2))
m.add(LSTM(units=50,return_sequences=True))
m.add(Dropout(0.2))
m.add(Dense(units=1))
m.compile(optimizer='adam',loss='mean_squared_error')
m.fit(x_train,y_train,batch_size=32,epochs=5)
p1=m.predict(x_train)
#======================================================================>
df1=pd.read_csv('C:\\vas.12344\\Google_Stock_Price_Test.csv')
df1
ff2=df1.iloc[:,1:2]
ff2=s1.transform(ff2)
#ff2.shape
dt=pd.concat((df['Open'],df1['Open']),axis=0)
inputs=dt[len(dt)-len(df1)-60:].values
#dt.shape
#as of now inputs are in the form of data frames, we should convert to numpy
#inputs=np.array(inputs)
inputs=inputs.reshape(-1,1)
inputs=s1.transform(inputs)
x_test=[]
for i in range(60,80):
   x_test.append(inputs[i-60:i,0])
x_test=np.array(x_test)
x_test=np.reshape(x_test,(x_test.shape[0],x_test.shape[1],1))
p2=m.predict(x_test)
#p2=s1.inverse_transform(p2)
#ff2=s1.inverse_transform(ff2)
#print(p1[0])
#print(p2[0])
#-----------------------------------------------------------
plt.plot(p2)
#plt.plot(ff2)
#plt.show()
print(p2)
```
