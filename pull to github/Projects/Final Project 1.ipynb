```python
#Credit Card Fraud Prediction Model
#Random Forest
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split,cross_val_score,GridSearchCV
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from feature_engine.outliers import Winsorizer
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier,GradientBoostingClassifier
from xgboost import XGBClassifier 
df=pd.read_csv('C:\\vas.12344\\card_transdata.csv')
#df.info()
df.describe()
df.isna().sum()
df.duplicated().sum()
f=Winsorizer(capping_method='iqr',tail='both',fold=1.5)

#sns.boxplot(df['distance_from_home'])
df['distance_from_home']=f.fit_transform(df[['distance_from_home']])
#sns.boxplot(df['distance_from_home'])
#sns.boxplot(df['distance_from_last_transaction'])
df['distance_from_last_transaction']=f.fit_transform(df[['distance_from_last_transaction']])
#sns.boxplot(df['distance_from_last_transaction'])
#sns.boxplot(df['ratio_to_median_purchase_price'])
df['ratio_to_median_purchase_price']=f.fit_transform(df[['ratio_to_median_purchase_price']])
#sns.boxplot(df['ratio_to_median_purchase_price'])
#sns.boxplot(df['repeat_retailer'])
#df['repeat_retailer']---------------->no outlier cleaning is possiable
#sns.boxplot(df['used_chip'])---------> for all remaining no Outlier cleaing possiable
#plt.show()


#df['repeat_retailer'],df['used_chip'],df['used_pin_number'],df['online_order'],df['fraud']
#df['distance_from_home'].skew()
df['distance_from_home']=df['distance_from_home']**(1/2)
df['distance_from_home'].skew()
#df['distance_from_last_transaction'].skew()
df['distance_from_last_transaction']=df['distance_from_last_transaction']**(1/2)
#df['distance_from_last_transaction'].skew()
df['ratio_to_median_purchase_price'].skew()
df['ratio_to_median_purchase_price']=df['ratio_to_median_purchase_price']**(1/2)
#df['ratio_to_median_purchase_price'].skew()
#df['repeat_retailer'].skew()---------->from here to end skewness os not possiable
df.corr()
#sns.heatmap(df.corr(),annot=True)
#no colinearity problem because no 2 features are corelated to each other
#plt.bar(df['distance_from_home'],df['fraud'])
#plt.show()
x=df.drop('fraud',axis=1)
y=df['fraud']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
m=RandomForestClassifier()
m.fit(x_train,y_train)
y1=m.predict(x_train)
y2=m.predict(x_test)
a1=accuracy_score(y_train,y1)
a2=accuracy_score(y_test,y2)
cv=cross_val_score(m,x,y,cv=5)
cv=cv.mean()
print(a1,a2,cv)
#------------------------------------------------------->Hyper perameteric tuning
#m1=RandomForestClassifier(random_state=0)
#v={'n_estimators':list(range(1,10))}
#gsv=GridSearchCV(m1,v,scoring='accuracy',cv=5)
#gsv.fit(x_train,y_train)
#gsv.best_params_
m2=RandomForestClassifier(n_estimators=8,random_state=0)
m2.fit(x_train,y_train)
yy1=m2.predict(x_train)
yy2=m2.predict(x_test)
aa1=accuracy_score(y_train,yy1)
aa2=accuracy_score(y_test,yy2)
cv2=cross_val_score(m2,x,y,cv=5)
cv2=cv.mean()
print(aa1,aa2,cv2)
#In this case HYT perameters is the best perameters
```
