```python
#Sentiment Analysis For Financial News
#ADA Bossting
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
import nltk
from nltk.stem import PorterStemmer
import re
from sklearn.preprocessing import LabelEncoder
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
from nltk.stem import PorterStemmer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split,cross_val_score
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier,GradientBoostingClassifier
from xgboost import XGBClassifier
ps=PorterStemmer()
#df=pd.read_csv('C:\\vas.12344\\Fiance sentiment analaysis.csv')
df=pd.read_csv('C:\\vas.12344\\sentiment analysis for financial news.csv')
#Sentiment
l=[]
for i in range(len(df)):
   r1=re.sub('[^a-zA-Z]'," ",df['Text'][i])
   r1=r1.lower()
   r1=r1.split()
   r1=[ps.stem(i) for i in r1 if not i in set(stopwords.words('english'))]
   r1=' '.join(r1)
   l.append(r1)

#m=CountVectorizer(max_features=1000)

m=CountVectorizer()
x=m.fit_transform(l).toarray()
#df['Sentiment']=pd.get_dummies(df['Sentiment'],drop_first=True)
#e1=LabelEncoder()
#df['Sentiment']=e1.fit_transform(df['Sentiment'])
#y=df['Sentiment']
#y=pd.get_dummies(df['Sentiment'],drop_first=True)
#y=np.array(y)
y=df['Sentiment']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
m1=AdaBoostClassifier()
m1.fit(x_train,y_train)
p1=m1.predict(x_train)
p2=m1.predict(x_test)
cv=cross_val_score(m1,x,y,cv=5)
cv=np.mean(cv)
a1=accuracy_score(y_train,p1)
a2=accuracy_score(y_test,p2)
a1,a2,cv
```
